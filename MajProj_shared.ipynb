```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
sa_df = pd.read_csv("shared_articles.csv")
sa_df.head(5)
```

```python
sa_df = sa_df[sa_df['eventType'] == 'CONTENT SHARED']
sa_df.head(5)
```

```python
interactions_df = pd.read_csv('users_interactions.csv')
interactions_df.head(10)
```

# shared_articles.csv

```python
event_type_strength = {
   'VIEW': 1.0,
   'LIKE': 2.0, 
   'BOOKMARK': 2.5, 
   'FOLLOW': 3.0,
   'COMMENT CREATED': 4.0,  
}
```

```python
interactions_df['eventStrength'] = interactions_df['eventType'].apply(lambda x: event_type_strength[x])
```

```python
interactions_df
```

Recommender systems have a problem known as user cold-start, due to the lack of information to model their preferences.For this reason, we are keeping in the dataset only users with at leas 5 interactions

```python
users_interactions_count_df = interactions_df.groupby(['personId','contentId']).size().groupby('personId').size()
print("Users : ",len(users_interactions_count_df))
```

```python
users_with_enough_interactions_df = users_interactions_count_df[users_interactions_count_df >= 5].reset_index()[['personId']]
```

```python
print("users with at least 5 interactions : ",len(users_with_enough_interactions_df))
```

```python
print('of interactions: %d' % len(interactions_df))
interactions_from_selected_users_df = interactions_df.merge(users_with_enough_interactions_df, 
               how = 'right',
               left_on = 'personId',
               right_on = 'personId')

interactions_from_selected_users_df
```

```python
print('of interactions from users with at least 5 interactions: %d' % len(interactions_from_selected_users_df))
```

```python
def smooth_user_preference(x):
    return math.log(1+x, 2)

interactions_full_df = interactions_from_selected_users_df \
                    .groupby(['personId', 'contentId'])['eventStrength'].sum() \
                    .apply(smooth_user_preference).reset_index()

print('of unique user/item interactions : ', len(interactions_full_df))

interactions_full_df.head(10)
```
